```python
# Creating a list with the number columns, and another with the text.
num = (movies.dtypes == float) | (movies.dtypes == int)
num
```

```python
# This fuction shows the colums labels or names.
num.index
```

```python
# In this line we show that this objects are iterable.
for element in num:
    print(element)
```

```python
# Same example as previous case, but with the index.
for element in num.index:
    print(element)
# So, we can see that num.index generates a iterable with index of our table.
```

```python
# This is a python magic method, which will create a list of c's and each 'c' will be the element.
# The element will be add to the list only if it verifies the condition 'num[element]'
num_columns = [element for element in num.index if num[element]]
num_columns
```

```python
# The same but with obj instead of num.
obj = movies.dtypes == object
obj_columns = [element for element in obj.index if obj[element]]
obj_columns
# These columns containing text type objects.
```

```python
num_columns
```

```python
# Here we are creating an array with only the numeric columns. 
movies_num = movies[num_columns]
movies_num
```
